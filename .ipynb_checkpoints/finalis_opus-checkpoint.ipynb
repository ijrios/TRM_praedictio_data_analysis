# Introducción

El mercado de divisas ha tenido un crecimiento significativo desde las últimas décadas. Debido a la internacionalización en empresas de todo el mundo, el tipo de cambio se considera un factor crucial en los principales negocios de todo el mundo. Para una persona común, el cambio de divisas entra en escena cuando la persona viaja a otro país por negocios o vacaciones. Las fluctuaciones del tipo de cambio no importan mucho ya que la cantidad sería menor. Pero cuando el comercio internacional toma lugar, la cantidad es muy grande y la pequeña diferencia, incluso el cambio decimal, en el tipo de cambio puede tener un efecto significativo. impacto. Considere un acuerdo comercial finalizado entre dos partes por un monto de mil millones de dólares comprador que reside en EE. UU. y vendedor residente en Colombia. Una ligera variación en el tipo de cambio puede tener un impacto significativo y es posible que el vendedor tenga que pagar más esa cantidad liquidada.

La Tasa de Cambio Representativa del Mercado (TCRM), corresponde al promedio ponderado de las operaciones de compra y venta de contado de dólares de los Estados Unidos de América a cambio de moneda legal colombiana. Este documento modela los tipos de cambio anuales entre USD/CO, y compara los datos reales con pronósticos desarrollados utilizando análisis de series de tiempo durante el período de 1992 a 2022. Los datos semanales oficiales del Banco Nacional de la República de Colombia se utilizan para el presente estudio.

El objetivo principal del proyecto es construir un sistema que pueda predecir el tipo de cambio para el día siguiente o cualquiera de los futuros día/s con la máxima precisión. Esto ayudará a las casas de cambio y las personas [naturales] a tener una visión anticipada del tipo de cambio de divisas antes de hacer cualquier cambio, tipo de transacción. Nuestro enfoque es realizar análisis de series de tiempo para el peso colombiano contra USD. Usando los modelos podemos predecir los valores futuros basados en los valores observados, dados en términos de datos de series temporales de valores de tipo de cambio de moneda de cada día.

### Estacionariedad

Recordemos que para usar ARIMA de forma adecuada, debemos tener una serie estacionaria. Usemos una prueba estadística para ver si la serie es o no es estacionaria.

La hipótesis nula  de esta prueba es que la serie no es estacionaria.


```python
from statsmodels.tsa.stattools import adfuller
print('Dickey-Fuller test:')
dftest=adfuller(ts_cop_weekly['PRETIUM'], autolag='AIC')
dfoutput=pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', 'Mora usus', 'No. per Obs'])
for key,value in dftest[4].items():
    dfoutput['Valores critici (%s)'%key] = value
print(dfoutput)
```

Vemos que el p-valor es de 0.98, evidentemente no se puede rechazar la hipótesis nula, por lo que no hay suficiente evidencia para decir que nuestra serie si sea estacionaria.

Por ende, miremos la serie en diferencias a ver si esta si es estacionaria.

Para llevarla a diferencias, usamos el método diff() de pandas.

```python
### Llevar la serie a diferencias
diff_data = ts_cop_weekly.diff().dropna()
diff_data
```

```python
plt.figure(figsize=(18,8))
plt.xlabel("Weekly")
plt.ylabel("Repraesentativas mercatus commutationem rate - Historical")
plt.plot(diff_data)
```

```python
print('Dickey-Fuller test:')
dftest=adfuller(diff_data['PRETIUM'], autolag='AIC')
dfoutput=pd.Series(dftest[0:4], index=['Test Statistics', 'p-value', 'Mora usus', 'No. per Obs'])
for key,value in dftest[4].items():
    dfoutput['Valores critici (%s)'%key] = value
print(dfoutput.round(5))
```

El p-value es menor a 0.05. El nivel de confianza es mayor a 99% entonces ya podemos concluir que la serie es estacionaria, lo cual es suficiente para este ejemplo.

Esto indica que la serie debe diferenciarse una vez para ser estacionaria. Esto hay que tenerlo en cuenta más adelante al construir nuestro modelo ARIMA.

### División de los datos en entrenamiento y prueba

Antes de continuar, vamos a aplicar un procesamiento en los datos que es muy común en problemas de analítica.

Lo que vamos a hacer es dividir nuestros datos en datos de entrenamiento y datos de prueba.

Esto se hace para que el modelo que construyamos sea entrenado y ajustado solamente con el subconjunto de los datos que son de entrenamiento. De esta forma, luego podemos evaluar el modelo con los datos de prueba para ver cómo es su rendimiento. Esto se hace para tener más similitudes con un comportamiento real: cuando vaya a usar el modelo en condiciones reales, lo usaré sobre datos que no existían y que el modelo jamás había visto. Por eso debo usarlo sobre los datos que deje aparte como conjunto de prueba, en vez de sobre los datos de entrenamiento.

En series de tiempo, este proceso se suele hacer de forma cronológica: simplemente asumo que una pequeña parte final de mi serie de tiempo serán datos de prueba, y el resto serán los de entrenamiento.

En este caso, asumiremos que el 70% inicial de la serie serán datos de entrenamiento, y el 30% restante serán los de prueba.


```python
# Scinditur in disciplina (train) et test (test) occidit
X = ts_cop_weekly.copy()
size = int(len(X) * 0.7)
train, test = X[0:size], X[size:len(X)]
```

Grafiquemos nuestros datos de entrenamiento y de prueba:

```python
### nos graph serie
plt.figure(figsize=(18,8))
plt.xlabel("Weekly")
plt.ylabel("Repraesentativas mercatus commutationem rate - Historical")
plt.plot(train, color='blue', label='Disciplina data')
plt.plot(test, color='orange', label='Test data')
plt.legend()
```

Así, nuestros datos de entrenamiento van desde 1992 hasta inicios del 2014, y de ahí en adelante estarán los datos de prueba. 
- Ahora sí estamos listos para empezar a construir modelos de series de tiempo para nuestra problemática.

# 4. Modelación 

### Construcción del modelo ARIMA

Para este proyecto, el primer modelo que construimos fue el modelo ARIMA (Promedio móvil integrado autorregresivo).  La técnica ARIMA  es un modelo estadístico que se utiliza para pronosticar el conjunto de datos que tiene una naturaleza de serie temporal. El modelo ARIMA no es más que una composición de tres modelos, a saber, autorregresión (AR), integración (I) y media móvil (MA).

Recordemos que los hiperparámetros de un modelo ARIMA son (p,d,q), donde p es para la parte "AR" (cuántos rezagos incluir), d es para la parte "I" (cuántas veces integrar o diferenciar la serie) y la q es para la parte "MA" (cuántos medias móviles del error se incluirán).

Observando detenidamente la gráfica, vemos que los datos de tipos de cambio de divisas entre USA y colombia no son estacionarios y esto es cierto para cualquier tipo de cambio de moneda. Para hacerlo estacionario se tuvo que diferenciar la serie, en el modelo ARIMA para aplicar este cambio la "d" será igual a 1. Nos falta determinar entonces valores para la p y la q.


```python
from statsmodels.tsa.arima.model import ARIMA

plt.figure(figsize=(18,8))

model=ARIMA(train, order=(2,1,2)) # p, d, q
results= model.fit()
plt.plot(train)

prediccion_arima =  results.fittedvalues[1:]
plt.plot(prediccion_arima, color='red')
print('plotting ARIMA model')
```

Este gráfico es más "conservador": intenta modelar muy bien la mayoría de los puntos, incluso en los picos altos.

De la misma forma, resulta valioso hacer un primer análisis con esta visualización.

Podemos por ejemplo ver que:

- Se tienen datos de forma mensual, desde 1992 a 2022, para el valor diario del precio del peso colombiano frente al dolar estadounidense.

- Se nota que hay una tendencia creciente clara en los datos, por lo tanto, no es estacionaria (pero se puee convertir).

- Se nota claramente que las observaciones están relacionadas entre sí, es decir, hay dependencia (común en una serie de tiempo).

- Hay outliers evidentes en la serie de datos, sin embargo, son congruentes, ya que le dólar en cualquier momento puede alcanzar un punto máximo y volver a caer, no afecta el modelado y predicción de datos.


```python
### Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

mse = mean_squared_error(train['PRETIUM'][1:], prediccion_arima)
mae = mean_absolute_error(train['PRETIUM'][1:], prediccion_arima)


print('MSE: '+str(mse))
print('MAE: '+str(mae))

#mse = mean_squared_error(y_true, y_pred)
```

```python
#devita quidam caveats
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)
```

### SARIMA

Claramente habíamos notado que la serie tiene una estacionalidad. Por dicha razón, incluyamos entonces el modelo SARIMA para los efectos estacionales.

Agreguemos dos rezagos a los efectos estacionales.

```python
plt.figure(figsize=(18,8))

model=ARIMA(train, order=(3,1,1), seasonal_order=(2, 1, 1,12)) 
results1=model.fit()
plt.plot(train)

prediccion_sarima = results1.fittedvalues[1:]

plt.plot(prediccion_sarima, color='red')
print('plotting SARIMA model')
```

- **Podemos observar que el modelo que más exacto dio, fue el ARIMA, por lo tanto seguiremos con este, sin embargo, sus resultados fueron muy parecidos, procedemos a realizar la predicción de los datos futuros.**

# 6. Despliegue (Predicciones a futuro)

### - Primer metodo para predecir  - One step ahead (un paso a la vez)
La de one-step ahead lo que hace es predecir de un mes a la vez: predigo lo de febrero sabiendo lo que paso en enero, lo de marzo sabiendo lo de febrero, lo de abril sabiendo lo de marzo, y así sucesivamente.

Esta predicción es la que se debería utilizar si voy a reejecutar cada mes el modelo.

```python
historia = list(train.values.copy())
praedictiones = list()
for t in range(len(test)):
	exemplar =  ARIMA(historia, order=(p_fin, d_fin, q_fin))
	exemplar_fit = exemplar.fit()
	output = exemplar_fit.forecast()
	yhat = output[0]
	praedictiones.append(yhat)
	obs = test.values[t]
	historia.append(obs)
#print('predicted=%f' % (yhat))

# Métricas en prueba
mae = mean_absolute_error(test, praedictiones)
print('Test MAE: %.3f' % mae)
mse = mean_squared_error(test, praedictiones)
print('Test MSE: %.3f' % mse)
```

**Claramente se ve que se obtienen un MAE y MSE que aún son buenos pero no tan buenos como cuando lo usabamos con datos con los que entrenamos, sin embargo, aún sigue siendo un buen modelo para predir datos, con los datos entrenados y utilizando el ARIMA el error absoluto medio más bajo (13 pesos colombianos), con un paso a la vez es mayor: 40 pesos colombianos**


```python
experimentus = pd.DataFrame()
experimentus['Pretium'] = praedictiones
experimentus.set_index(test.index, inplace=True)
experimentus
```

#### One step ahead (un paso a la vez - Gráfica)

```python
### Predicciones a futuro
plt.figure(figsize=(18,8))
plt.plot(train, label='Verus data')

## Forecast future
plt.plot(experimentus, color='red', label='Predicción en la muestra One-step')
plt.legend()
```

### Segundo metodo para predicir - Multi step ahead (Todos los períodos de una sola vez)

Por otro lado, si voy a ejecutar el modelo por ejemplo solo una vez cada año, entonces debe predecir los siguientes X meses a partir de hoy, sin tener más información adicional.

```python
## Predicciones Multi-step ahead 
historia_1 = list(train.values.copy())
praedictiones_1 = list()

model_1 =  ARIMA(historia_1, order=(p_fin, d_fin, q_fin))
model_fit_1 = model_1.fit()
output_1 = model_fit_1.forecast(len(test))   ### Esta es la linea clave: predigo todo a la vez, por eso le paso "len(test)""
yhat_1 = output_1
praedictiones_1 = yhat_1

mae = mean_absolute_error(test, praedictiones_1)
print('Test MAE: %.3f' % mae)
mse = mean_squared_error(test, praedictiones_1)
print('Test MSE: %.3f' % mse)
```

Aquí obviamente al vernos obligados a predecir mucho más a futuro, nuestras métricas se ven mucho más afectadas. El 
error absoluto medio es muy alto y supera los 1174 pesos colombianos.

- Esto tiene sentido con nuestra lógica: es más fácil predecir por ejemplo el precio del dólar de mañana que el precio del dolar dentro de 20 días.
- Pero con este modelo vemos que los datos se alejan demasiado del valor real, no es tan preciso, por lo tanto, no se usará para la predicción a futuro.

```python
experimentus_1 = pd.DataFrame()
experimentus_1['Pretium'] = praedictiones_1
experimentus_1.set_index(test.index, inplace=True)
experimentus_1
```

### Tercer metodo para predicir a *futuro*

Con el siguiente código, generamos a futuro las predicciones del modelo ARIMA, con un intevalo de confianza del 95%, para los siguientes  meses a futuro:

```python
### Modelo entrenado con todos los datos
model =  ARIMA(ts_cop_weekly, order=(p_fin, d_fin, q_fin)) #
# Mirar que pongo data para usar todos los datos
results=model.fit()
prediccion_sarima = results.fittedvalues[1:]

### Predecir 48 pasos (meses) a futuro
x=results.get_forecast(steps=48)

### El valor predicho es
x_pred= x.predicted_mean
pd.DataFrame(x_pred)
```

- Como podemos observar los valores que se predicen a futuro son muy parecidos, esta serie es muy difícil de modelar, los datos tienden a tener una tendencia creciente, además los datos varían mucho, el precio del peso colombiano con respecto al dólar depende de muchos factores externos que pueden cambiar día a día.

```python
### Y su intervalo de confianza es
x_ci = x.conf_int()
x_ci.head()
```

```python
### Predicciones a futuro
plt.figure(figsize=(18,8))

plt.plot(ts_cop_weekly, label='Real data')

prediccion_sarima = results.fittedvalues[1:]

plt.plot(prediccion_sarima, color='red', label='In-sample prediction')

## Forecast future
plt.plot(x_pred, color='green', label='Forecast')
plt.plot(x_ci['lower PRETIUM'], color='yellow', label='Lower CI')
plt.plot(x_ci['upper PRETIUM'], color='purple', label='Upper CI')

plt.legend()
```

### Predicciones reejecutadas en un día futuro

En un proyecto real debemos tener un código listo para reentrenar el modelo con los datos actualizados y generar las predicciones a partir de la fecha final de los datos actualizados, por lo tanto, organizamos el modelado para datos que se actualizan constantemente.

```python
### Read updated data cum novis dies
data_new =pd.read_csv('data/dolar.csv')
data_new['DATE']=pd.to_datetime(data_new['VALIDITATEM_EX'], infer_datetime_format=True)
data_new=data_new.set_index(['DATE'])
data_new = data_new.drop(['VALIDITATEM_TO'], axis=1)
data_new = data_new.drop(['VALIDITATEM_EX'], axis=1)
data_new = data_new.drop(['UNITAS'], axis=1)
data_new = data_new.sort_index() 
data_new = data_new.resample('W').mean()
data_new
```

```python
### Updated praedictiones
model =  ARIMA(data_new, order=(p_fin, d_fin, q_fin))
results=model.fit()

x=results.get_forecast(steps=48)
x_pred= x.predicted_mean
x_ci = x.conf_int()
```

```python
### Ploting
plt.figure(figsize=(20,6))

plt.plot(data_new, label='Real data')

prediccion_sarima = results.fittedvalues[1:]

plt.plot(prediccion_sarima, color='red', label='In-sample prediction')


plt.plot(x_pred, color='green', label='Forecast')
plt.plot(x_ci['lower PRETIUM'], color='yellow', label='Lower CI')
plt.plot(x_ci['upper PRETIUM'], color='purple', label='Upper CI')

plt.legend()
```
