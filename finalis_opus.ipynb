### Estacionariedad

Recordemos que para usar ARIMA de forma adecuada, debemos tener una serie estacionaria. Usemos una prueba estadística para ver si la serie es o no es estacionaria.

La hipótesis nula  de esta prueba es que la serie no es estacionaria.


```python
from statsmodels.tsa.stattools import adfuller
print('Prueba de Dickey-Fuller:')
dftest=adfuller(ts_cop_semana['VALOR'], autolag='AIC')
dfoutput=pd.Series(dftest[0:4], index=['Estadística de prueba', 'valor p', 'Retrasos utilizados', 'No. de Obs'])
for key,value in dftest[4].items():
    dfoutput['Valores criticos (%s)'%key] = value
print(dfoutput)
```

Vemos que el p-valor es de 0.98, evidentemente no se puede rechazar la hipótesis nula, por lo que no hay suficiente evidencia para decir que nuestra serie si sea estacionaria.

Por ende, miremos la serie en diferencias a ver si esta si es estacionaria.

Para llevarla a diferencias, se usa el método diff() de pandas.

```python
### Llevar la serie a diferencias
diff_data = ts_cop_semana.diff().dropna()
diff_data
```

```python
plt.figure(figsize=(20,10))
plt.xlabel("Mensual")
plt.ylabel("Tasa de Cambio Representativa del Mercado (diferencias)")
plt.plot(diff_data)
```

```python
print('Prueba de Dickey-Fuller:')
dftest=adfuller(diff_data['VALOR'], autolag='AIC')
dfoutput=pd.Series(dftest[0:4], index=['Estadística de prueba', 'valor p', 'Retrasos utilizados', 'No. de Obs'])
for key,value in dftest[4].items():
    dfoutput['Valores criticos (%s)'%key] = value
print(dfoutput.round(5))
```

El p-valor es menor a 0.05. El nivel de confianza es mayor a 99% ya podemos concluir que la serie si es estacionaria, lo cual es suficiente para este ejemplo.

Esto indica que la serie debe diferenciarse una vez para ser estacionaria. Esto hay que tenerlo en cuenta más adelante al construir nuestro modelo ARIMA.

### División de los datos en entrenamiento y prueba

Antes de continuar, vamos a aplicar un procesamiento en los datos que es muy común en problemas de analítica.

Lo que vamos a hacer es dividir nuestros datos en datos de entrenamiento y datos de prueba.

Esto se hace para que el modelo que construyamos sea entrenado y ajustado solamente con el subconjunto de los datos que son de entrenamiento. De esta forma, luego podemos evaluar el modelo con los datos de prueba para ver cómo es su rendimiento. Esto se hace para tener más similitudes con un comportamiento real: cuando vaya a usar el modelo en condiciones reales, lo usaré sobre datos que no existían y que el modelo jamás había visto. Por eso debo usarlo sobre los datos que deje aparte como conjunto de prueba, en vez de sobre los datos de entrenamiento.

En series de tiempo, este proceso se suele hacer de forma cronológica: simplemente asumo que una pequeña parte final de mi serie de tiempo serán datos de prueba, y el resto serán los de entrenamiento.

En este caso, asumiré que el 90% inicial de la serie serán datos de entrenamiento, y el 10% restante serán los de prueba.


```python
# Dividir en conjuntos de entrenamiento (train) y prueba (test)
X = ts_cop_semana.copy()
size = int(len(X) * 0.9)
train, test = X[0:size], X[size:len(X)]
```

Pintemos nuestros datos de entrenamiento y de prueba:

```python
### Pintar la serie
plt.figure(figsize=(20,10))
plt.xlabel("Mensual")
plt.ylabel("Tasa de Cambio Representativa del Mercado -Historico")
plt.plot(train, color='blue', label='Datos de entrenamiento')
plt.plot(test, color='orange', label='Datos de prueba')
plt.legend()
```

Así, nuestros datos de entrenamiento van desde 1992 hasta inicios del 2020, y de ahí en adelante estarán los datos de prueba. 
- Ahora sí estamos listos para empezar a construir modelos de series de tiempo para nuestra problemática.

# 4. Modelación 

### Construcción del modelo ARIMA

Recordemos que los hiperparámetros de un modelo ARIMA son (p,d,q), donde p es para la parte "AR" (cuántos rezagos incluir), d es para la parte "I" (cuántas veces integrar o diferenciar la serie) y la q es para la parte "MA" (cuántos medias móviles del error se incluirán).

Cómo ya vimos que luego de diferenciar una vez la serie era estacionaria, ya sabemos que nuestra "d" será igual a 1. Nos falta determinar entonces valores para la p y la q.


```python
from statsmodels.tsa.arima.model import ARIMA

plt.figure(figsize=(20,10))

model=ARIMA(train, order=(2,1,2)) # p, d, q
results= model.fit()
plt.plot(train)

prediccion_arima =  results.fittedvalues[1:]
plt.plot(prediccion_arima, color='red')
print('plotting ARIMA model')
```

Nótese que este gráfico parece más "conservador": intenta modelar muy bien la mayoría de los puntos, a pesar de que en los picos no suba tanto como los otros escenarios.

```python
### Metricas
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

mse = mean_squared_error(train['VALOR'][1:], prediccion_arima)
mae = mean_absolute_error(train['VALOR'][1:], prediccion_arima)


print('MSE: '+str(mse))
print('MAE: '+str(mae))

#mse = mean_squared_error(y_true, y_pred)
```

```python
#Para evitar algunas advertencias
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning
warnings.simplefilter('ignore', ConvergenceWarning)
```

### SARIMA

Claramente habíamos notado que la serie tiene una estacionalidad. Por dicha razón, incluyamos entonces el modelo SARIMA para los efectos estacionales.

Agreguemos dos rezagos a los efectos estacionales.

```python
plt.figure(figsize=(20,10))

model=ARIMA(train, order=(3,1,1), seasonal_order=(2, 1, 1,12)) 
results1=model.fit()
plt.plot(train)

prediccion_sarima = results1.fittedvalues[1:]

plt.plot(prediccion_sarima, color='red')
print('plotting SARIMA model')
```

De la misma forma, resulta valioso hacer un primer análisis con esta visualización.

Podemos por ejemplo ver que:

- Se tienen datos de forma mensual, desde 1992 a 2022, para el valor diario del precio del peso colombiano con respecto al dolar

- Se nota que hay una tendencia creciente clara en los datos (no es estacionaria)

- Se nota claramente que las observaciones están relacionadas entre si, es decir, hay dependencia (común en una serie de tiempo)

- No hay outliers evidentes en la serie de datos (ni puntos individuales outliers ni grupos de puntos outliers)


```python
### Metrica
mse1 = mean_squared_error(train['VALOR'][1:], prediccion_sarima)
mae1 = mean_absolute_error(train['VALOR'][1:], prediccion_sarima)

print('MSE: '+str(mse1))
print('MAE: '+str(mae1))

#mse = mean_squared_error(y_true, y_pred)
```

##### One step ahead (un paso a la vez)
La de one-step ahead lo que hace es predecir de un mes a la vez: predigo lo de febrero sabiendo lo que paso en enero, lo de marzo sabiendo lo de febrero, lo de abril sabiendo lo de marzo, y así sucesivamente.

Esta predicción es la que se debería utilizar si voy a reejecutar cada mes el modelo.

```python
historia = list(train.values.copy())
praedictiones = list()
for t in range(len(test)):
	exemplar =  ARIMA(historia, order=(p_fin, d_fin, q_fin))
	exemplar_fit = exemplar.fit()
	output = exemplar_fit.forecast()
	yhat = output[0]
	praedictiones.append(yhat)
	obs = test.values[t]
	historia.append(obs)
#print('predicted=%f' % (yhat))

# Métricas en prueba
mae = mean_absolute_error(test, praedictiones)
print('Test MAE: %.3f' % mae)
mse = mean_squared_error(test, praedictiones)
print('Test MSE: %.3f' % mse)
```
